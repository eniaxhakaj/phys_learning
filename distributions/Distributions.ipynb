# Gamma

```python
test_x = np.linspace(0, 60, num=5000)
_, ax = plt.subplots()
for a, s in[(5, 2), (5, 3), (10, 2), (10, 3)]:
    ax.plot(test_x, gamma(a=a, scale=s).pdf(test_x), label="shape: {}, scale: {}".format(a, s))

ax.axvline(27, color="red", label="Peak at 27")
ax.legend()
```
